# 话题1：日常生活中的等差数列和等比数列

## 城市中交通路线的平均长度 -- 等差数列与堆垒求和

假设城市中： 道路呈现为一个$n$*$m$ （$n$行$m$列）的方形网络，网络中的每个节点为一个公交站点，相邻的站点的间距为$L$:

问题： 从某一个站点到另一个站点的平均长度为多少？ 不允许走回头路，也不允许兜圈子，这样从一个站点到另一个站点的不同走法就有了相同的路程，于是就可以假设两个站点之间的行程中最多转弯一次。

先考虑在一行的情况，每一行有$m$列（$m$个站点），每一个都可能是起始点和结束点。  

设： 从第$i$个站点出发向东（右）走，可能经过的的路程有： $1$, $2$, $3$,$\dotsi$,$m-i$个$L$。则从$i$站出发向右所有可能走的站的路程之和为:
$$(1+2+3+\dotsi+(m-i))L = \frac{(1+m-i)(m-i)}{2} L$$

考虑不同的始发站$i=1,2,\dotsi,m$,可以得到一行中所有向右的路程总和.

<a name="formula-1"></a>
$$
\begin{align}
\tag{1}
L\cdot\sum_{i=1}^m\frac{(1+m-i)(m-i)}{2} &= \frac{1}{2}\sum_{i=1}^m(m-i+m^2-2mi+i^2)L \\
&= \frac{1}{2}\sum_{i=1}^m(m^2+m + i^2 - (2m+1)i)L \\
&= \frac{1}{2}\Bigg( (m^2+m)m + \sum_{i=1}^m(i^2 - (2m+1)i)\Bigg)L \\
&= \frac{1}{2}\Bigg( (m^2+m)m - (2m+1)\sum_{i=1}^mi + \sum_{i=1}^mi^2\Bigg)L \\
\end{align}
$$

其中:$\displaystyle\sum_{i=1}^mi = \frac{(m+1)m}{2}$ , 而在以下情况中:

\begin{cases}
   (m+1)^3 = m^3 + 3m^2+ 3m+1 \\
   m^3 = (m-1)^3 + 3(m-1)^2+ 3(m-1)+1 \\
   (m-1)^3 = (m-2)^3 + 3(m-2)^2+ 3(m-2)+1 \\
   \vdots \\
   3^3 = 2^3 + 3\times2^2 + 3\times2+1 \\
   2^3 = 1^3 + 3\times1^2 + 3\times1+1
\end{cases}

一次将下方的式子带入到上方的式子中,例如: 
$$ 
\begin{equation}
\begin{split}   
3^3 &= 2^3 + 3\times2^2 + 3\times2+1 \\
  &= 1^3 + 3\times1^2 + 3\times1+1 + 3\times2^2 + 3\times2+1 \\
  &= 1^3 + 3\cdot(1^2+2^2) + 3\cdot(1+2) + 2
\end{split}
\end{equation} 
$$

以此类推,可得: 
$$ 
\begin{equation}
\begin{split}  
(m+1)^3 &= 1^3 + 3\cdot(1^2+2^2+\dotsi+m^2)+ 3(1+2+\dotsi+m)+m \\
        &= 1^3 + 3\cdot\sum_{i=1}^mi^2+3\cdot\sum_{i=1}^mi+m 
\end{split}
\end{equation} $$

有上述式子可以看出: $  \displaystyle\sum_{i=1}^mi^2 = \frac{(m+1)^3 - 1^3 - 3\cdot\displaystyle\sum_{i=1}^mi - m}{3} $

代入$\displaystyle\sum_{i=1}^mi = \frac{(m+1)m}{2}$ 并简化: 

$$
\begin{equation}
\begin{split} 
\sum_{i=1}^mi^2 &= \frac{(m+1)^3 - 1^3 - 3\cdot\frac{(m+1)m}{2}  - m}{3} \\
&= \frac{2(m+1)^3 - 2 - 3m(m+1)  - 2m}{6} \\
&= \frac{2(m+1)^3 - 3m(m+1)-2(m+1)}{6} \\
&= \frac{2(m+1)^3 - (3m+2)(m+1)}{6} \\
&= \frac{(2(m^2 + 1 + 2m) - 3m-2)(m+1)}{6} \\
&= \frac{(2m^2 + 2 + 4m - 3m-2)(m+1)}{6} \\
&= \frac{(2m^2 + m)(m+1)}{6} \\
&= \frac{m(2m + 1)(m+1)}{6} \\
&= \frac{(2m^3 + 3m^2 + m)}{6} 
\end{split}
\end{equation} 
$$

得到 $$ \displaystyle\sum_{i=1}^mi^2 = \frac{(2m+1)(m+1)m}{6} $$ 代入[公式1](#formula-1)并简化:

$$
\begin{align}
L\cdot\sum_{i=1}^m\frac{(1+m-i)(m-i)}{2} &= \frac{1}{2}\Bigg( (m^2+m)m - (2m+1)\sum_{i=1}^mi + \sum_{i=1}^mi^2\Bigg)L \\
&= \frac{1}{2}\Bigg( (m^2+m)m - (2m+1)\frac{(m+1)m}{2} + \frac{(2m+1)(m+1)m}{6}\Bigg)L \\ 
&= \frac{1}{12}\Bigg( 6(m^2+m)m - 3(2m+1)(m+1)m + (2m+1)(m+1)m \Bigg)L \\
&= \frac{1}{12}\Bigg( 6m^3+6m^2 - 3(2m+1)(m+1)m + (2m+1)(m+1)m \Bigg)L \\
&= \frac{1}{12}\Bigg( 6m^3+6m^2 - 3(2m+1)(m+1)m + (2m+1)(m+1)m \Bigg)L \\
&= \frac{1}{6}\Bigg( 3(m^2+m)m - (2m+1)(m+1)m \Bigg)L \\
&= \frac{1}{6}\Bigg( 3m^3+3m^2  - ( 2m^3 + 3m^2 +m ) \Bigg)L \\
&= \frac{1}{6}( m^3  - m)L \\
&= \frac{1}{6}(m(m^2  - 1))L \\
\end{align}
$$

但是由于不仅可以向右走,也可以向左走,那么任意一行中可能的路程总和为: $$\frac{m(m^2 - 1)}{3}L $$

所以再只考虑单行的情况下, 对于$n$行的路程总和为: $$ \frac{nm(m^2 - 1)}{3}L $$

再加上对$n$行取两个的情况, 因为是需要考虑顺序的,所以我们采用排列的形式: 
$$ P_n^2 = P(n,2) = \frac{n!}{(n-2)!} = n(n-1)  $$

而$n$行任取两个从左到右路程和为: 
$$
 \frac{n(n-1)m(m^2 - 1)}{3}L
$$

所以, n行两两配对共有$n^2$种可能,于是 $n\times m$方形网络东西(左右或者说横)向所有可能的路程总和为: 
$$ \frac{n^2m(m^2 - 1)}{3}L $$

南北(上下)方向的计算是类似的,只不过是将$n$ 和$m$调换, 所以垂直的路程总合位:
$$
\frac{m^2n(n^2-1)}{3}L
$$

所以从任一站点出发到所有的站点总的路程为:  横向的总和 $+$ 竖向的总和: 
$$
\begin{align}
总长度 &= \frac{n^2m(m^2-1)}{3}L + \frac{m^2n(n^2-1)}{3}L 
\end{align}
$$

而从任一站点到其他的站点,总共有: $mn - 1$个选择,而总共有$mn$个站点,所以平均路程为:
$$ 
\begin{align}
avg &= \frac{n^2m(m^2-1)+ m^2n(n^2-1)}{3mn\cdot(mn-1)}L \\
&= \frac{n^2m^3-n^2m+ m^2n^3-m^2n}{3mn\cdot(mn-1)}L \\
&= \frac{n^2m^3-m^2n+ m^2n^3-n^2m}{3mn\cdot(mn-1)}L \\
&= \frac{nm^2(mn-1)+ n^2m(mn-1)}{3mn\cdot(mn-1)}L \\
&= \frac{(m+ n)}{3}L \\
\end{align}
$$

## 估计商品的价位 -- 等比中项与几何平均
一般地,给出两个正数 $a,b$,它们的几何平均数为$\sqrt{ab}$ ,也就是两者的等比中项,这是因为: 
$$
\frac{b}{\sqrt{ab}} = \frac{\sqrt{b}}{\sqrt{a}} = \frac{\sqrt{ab}}{a}
$$

关于等比中项和算术平均数(等差中项)有不等式:
$$
\sqrt{ab} \leqslant \frac{a+b}{2}
$$

两者(等比中项和算术平均数)之差为: 
$$
\frac{a+b}{2} - \sqrt{ab} = \frac{1}{2}(\sqrt{a}^2+\sqrt{b}^2 - 2\sqrt{ab}) = \frac{1}{2}(\sqrt{a} - \sqrt{b})^2
$$

所以,只有当$a = b$的时候,等比中项和算术平均数相等.

使用几何平均数的另一个好处在于,即使对于最大值和最小值的估计,也并非那样准确.实际上如果我们将几何平均数$\sqrt{ab}$看作以$b$为变量的函数$f(x) = \sqrt{ax}$, 则函数值$f(x)$的相对变化率$\frac{\Delta f(x))}{f(x)}$与自变量的相对变化率$\frac{\Delta x}{x}$的比值可以反映$x$的单位浮动对$f(x)$造成的影响, 这里考虑相对变化率而非绝对变化值是为了消除数量级和单位的影响, 计算可得:
$$
\frac{\Delta f(x)}{f(x)} \Big/ \frac{\Delta x}{x} = \frac{\Delta f}{\Delta x} \cdot \frac{x}{f(x)}
$$

通过求导链式法则,可以设$g(x) = ax $ , $h(x) = \sqrt{x} = x^\frac{1}{2} $,则$f(x) = h(g(x))$的导数为: $f'(x) = h'(g(x)) *g'(x) $

已知,$g'(x) = a , h'(x) = \frac{1}{2}x^{-\frac{1}{2}} $ 所以, $$
\begin{align}
f'(x) &= \frac{a}{2}(ax^{-\frac{1}{2}}) \\
 &= \frac{a}{2(ax^{\frac{1}{2}})} \\ 
 &=  \frac{a}{2\sqrt{ax}} 
\end{align}
$$

```python
from sympy import Symbol, diff,sqrt

x = Symbol('x')
a = Symbol('a')

y = sqrt(a*x)

derivative = diff(y, x)
derivative
```

将$f(x) 与f'(x)$都代入可得: 
$$ 
\begin{align}
\frac{\Delta f(x)}{f(x)} \Big/ \frac{\Delta x}{x} &= \frac{\Delta f}{\Delta x} \cdot \frac{x}{f(x)} \\
&= \frac{a}{2\sqrt{ax}}\cdot\frac{x}{\sqrt{ax}} \\
&= \frac{ax}{2ax} \\
&= \frac{1}{2}
\end{align}
$$

这意味着当自变量$x$相对变化 $2n%$ 时,函数值$f(x)$仅会发生$n%$左右的相对变化.

但是什么时候用算术平均数(等差中项) ,什么时候用几何平均数(等比中项)?可以借助离散随机变量的期望来分析.   
假设有关于离散型随机变量$X$的分比例,  其中 $p \in (0,1), q=1-p , x_1和x_2$两个相异正数,$\{x_1,x_2\}$为构成随机变量X的可能取值的集合, 则$X$的期望为: 
$$
E(X) = x_1p + x_2q
$$

### 等差中项
如果$E(X) = \frac{x_1+x_2}{2}$,由于$q = 1- p$则有: $$
x_1p + x_2q = \frac{x_1+x_2}{2} \\ 
x_1(p - \frac{1}{2}) + x_2(q - \frac{1}{2}) = 0 \\ 
x_1(p - \frac{1}{2}) + x_2(1 - p - \frac{1}{2}) = 0 \\ 
(x_1- x_2)(p - \frac{1}{2}) = 0 \\ 
$$
由于$x_1\ne x_2$,所以 $p=q=\frac{1}{2}$,这意味着随机变量$X$ 服从均匀分布.


### 等比中项
如果$E(X) = \sqrt{x_1x_2}$,由于$q = 1- p$则有: $$
x_1p + x_2q = \sqrt{x_1x_2} \\ 
p(\sqrt{x_1})^2 + (1-p)(\sqrt{x_2})^2 - \sqrt{x_1x_2} = 0 \\
p(\sqrt{x_1})^2  - p(\sqrt{x_2})^2 + (\sqrt{x_2})^2- \sqrt{x_1x_2} = 0 \\
p(\sqrt{x_1} + \sqrt{x_2})(\sqrt{x_1} - \sqrt{x_2}) + \sqrt{x_2}(\sqrt{x_2} - \sqrt{x_1}) = 0 \\
(p(\sqrt{x_1} + \sqrt{x_2}) - \sqrt{x_2})(\sqrt{x_2} - \sqrt{x_1}) = 0
$$
由于$x_1\ne x_2$,所以 $$p=\frac{\sqrt{x_2}}{\sqrt{x_1} + \sqrt{x_2}} , q =\frac{\sqrt{x_1}}{\sqrt{x_1} + \sqrt{x_2}} $$,即$p:q = \sqrt{x_2}:\sqrt{x_1}$ 


上面的分析说明,如果随机变量$X$的分布列更接近均匀分布,则期望$E(X)$越接近$x_1$和$x_2$的算术平均数; 如果随机变量$X$的分布列更接近右偏分布,且取较大值的概率小,取较小值的概率大,则期望$E(X)$更接近$x_1$和$x_2$的等比中项(几何平均数).

## 住院病人给药时间表的设计 -- 等比数列求和与极限控制法
假设病人在住院期间需要长期规律地注释针剂,(假设药物经针剂注射后直接释放到血液中,使血液中药物浓度,即血药浓度瞬间提升). 如果血液浓度超过$H$,则会中毒, 如果低于$L$,则不起效果.已知药物每次的注射量为定值,每次注射都会在当前的基础上增加$C_0$,根据渗透压原理,血药浓度越高,下降速度越快. 设$t$小时血药浓度为$C(t)$,如果不再给药,则经过$T$小时后将衰减为$C(t)\cdot e^{-\lambda T}$,其中$\lambda > 0$ 表示吸收率, $T > 0$ 表示,设第$k$次给药后的血液浓度为$R_k$, 第$k+1$次给药钱的药物浓度为$Q_k$,则:
$$
\begin{cases}
R_{k+1} = Q_k + C_0 \\
Q_{k+1} = R_{k+1} \cdot e^{-\lambda T}\quad k=1,2,3,\dots  \\
R_1=C_0 \\
\end{cases}
$$

于是可得: 
$$
\begin{align}
R_1 &= C_0 \\ 
R_2 &= C_0 + C_0\cdot e^{-\lambda T} \\
R_2 &= C_0 + C_0\cdot e^{-\lambda T}+ C_0\cdot e^{-2\lambda T} \\
\vdots \\
R_n &= C_0 + C_0\cdot e^{-\lambda T} + \dotsi + C_0\cdot e^{-(n-1))\lambda T} \\
\end{align}
$$

等比队列的求和公式为:  
$$
S = a\cdot\frac{1-r^n}{1-r}
$$  其中$a$ 是数列的第一项, $r$ 是数列的公比, $n$是数列的项数. 而在上面的$R_n$ 中,第一项为: $C_0$ , 公比为: $e^{-\lambda T}$,项数为: $n$ ,则:
$$
\begin{align}
R_n &= C_0\cdot \frac{1-e^{-n\lambda T}}{1-e^{-\lambda T}} \\ 
Q_n &= R_n\cdot e^{-\lambda T} \\
&=  C_0\cdot \frac{1-e^{-n\lambda T}}{1-e^{-\lambda T}} \cdot e^{-\lambda T}  \\ 
&=  C_0\cdot \frac{e^{-\lambda T} -e^{-(n+1)\lambda T}}{1-e^{-\lambda T}} \\ 
\end{align}
$$ 

令$n\to +\infty$ ,当n趋向于$+\infty$时,则 $1-e^{-n\lambda T}$ 趋向于1. 则可以算出$R_n$的极限为, : 
$$
\lim_{n\to\infty} R_n= \frac{C_0}{1-e^{-\lambda T}} \\
\lim_{n\to\infty} Q_n= \frac{C_0\cdot e^{-\lambda T} }{1-e^{-\lambda T}}
$$

而其中$R_n$的极限给药后的血液浓度, $Q_n$的极限则表示给药前的血液浓度.按照血药浓度阈值$H$和$L$ 即: 
$$
\begin{cases}
\frac{C_0}{1-e^{-\lambda T}} = H \\
\frac{C_0\cdot e^{-\lambda T} }{1-e^{-\lambda T}} = L
\end{cases}
$$

则时间T的计算过程为: 
$$
\begin{align}
\frac{H}{L} &= \frac { \frac{C_0}{1-e^{-\lambda T}} } {\frac{C_0\cdot e^{-\lambda T} }{1-e^{-\lambda T}}} \\ 
&= \frac{1}{ e^{-\lambda T} } \\
&= e^{\lambda T}  
\end{align}
$$ 简化后,可得$ln(\frac{H}{L}) = \lambda T $ , 则$T$为: 
$$
\begin{align}\
T &= \frac{ln(\frac{H}{L})}{\lambda} \\ 
  &= \frac{ln{H} - ln{L}}{\lambda}
\end{align}
$$

# 话题2: 二次和三次函数样条与数据的插值

多项式函数是形如: $$ p(x) = a_nx^n+a_{n-1}x^{n-1}+\dotsi+a_1x+a_0 $$的函数,其中$a_n,a_{n_1},\dots,a_1,a_0 \in \mathbb{R}, n \in \mathbb{N}^* , a_n \ne 0. 当n=1时, p(x) 为一次函数;当n=2时,p(x)为二次函数; 当n=3时,p(x)为三次函数$

所谓"插值",就是在原有数据点之间的"空挡"处,添加符合规律的"模拟数据". 假设平面上有3个数据点 $P_1(x_1,y_1),P_2(x_2,y_2),P_3(x_3,y_3)$ ,不妨设$x_1<x_2 < x_3$. 我们先取一个由两段"至多两次函数"拼接而成的函数: 
$$
f(x) =
\begin{cases}
S_1(x), x \in [x_1 , x_2] \\
S_2(x), x \in [x_2 , x_3] \\
\end{cases}
$$ 其中 $S_1(x) = a_1x^2+ b_1x+c_1,S_2(x) = a_2x^2+ b_2x+c_2 ( a_i , b_i,c_i \in \mathbb{R}, i = 1,2  )  ,$使得  : 
- $f(x_k) = y_k, k = 1,2,3 $使得$f(x)$经过每一个数据点;
- $f(x)$ 在$P_2$点处连续;
- $f(x)$ 在$P_2$点处存在一阶导数;
- $f(x)$ 在$P_2$点处存在二阶导数;   
由这些条件可得: 
$$
\begin{cases}

S_2(x_3) = y_3 \\
S'_1(x_2) = S'_2(x_2) \\
S''_1(x_2) = S''_2(x_2) \\
\end{cases}

```python
#imports
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import CubicSpline, interp1d
plt.rcParams['figure.figsize'] =(12,8)

x = np.linspace(0,2* np.pi)
y = np.sin(x)
# apply cubic spline interpolation
cs = CubicSpline(x, y)
# Apply Linear interpolation
linear_int = interp1d(x,y)

xs = np.linspace(0,2* np.pi)
ys = linear_int(xs)

# plot linear interpolation
plt.plot(x, y, 'o', label='data')
plt.plot(xs,ys, label="S", color='green')
plt.legend(loc='upper right', ncol=2)
plt.title('Linear Interpolation')
plt.show()

# plot cubic spline interpolation
plt.plot(x, y, 'o', label='data')
plt.plot(xs, np.sin(x), label='true')
plt.plot(xs, cs(xs), label="S")
plt.plot(xs, cs(xs, np.pi/2), label="S'")
plt.plot(xs, cs(xs, np.pi), label="S''")
plt.plot(xs, cs(xs, np.pi/2*3), label="S'''")
plt.ylim(-1.5, 1.5)
plt.legend(loc='upper right', ncol=2)
plt.title('Cubic Spline Interpolation')
plt.show()
```
